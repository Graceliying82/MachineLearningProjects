```python
import pandas as pd
import numpy as np
```

```python
import glob
import os
```

```python
testPaths = glob.glob("./testDigits/*.txt")
#Test data path
trainPaths = glob.glob("./trainingDigits/*.txt")
#Training Data Path
```

```python
digitArrayList = []
#use digitArrayList as X. Each element in digitArrayList is one digital file for a number

outPut=[]
#use outPut as Y. Each element in outPut indicate what number the digitArray is.

for path in testPaths:
    data =  pd.read_csv(path, header = None)
    digitArray = data.iloc[:,0].as_matrix()
    digitArrayList.append(digitArray)
    #add each X into digitArrayList
    

    filename = os.path.basename(path).split('.')[0]
    #add the no extension name to filename
    outPut.append(filename[0]) 
    
```

```python
testDigitArrayNP = np.array(digitArrayList)
#This is test X
print testDigitArrayNP.shape

testOutPutNP = np.array(outPut)
#This is test Y
```

```python
digitArrayList = []
#use digitArrayList as X. Each element in digitArrayList is one digital file for a number

outPut=[]
#use outPut as Y. Each element in outPut indicate what number the digitArray is.

for path in trainPaths:
    data =  pd.read_csv(path, header = None)
    digitArray = data.iloc[:,0].as_matrix()
    digitArrayList.append(digitArray)
    #add each X into digitArrayList
    

    filename = os.path.basename(path).split('.')[0]
    #add the no extension name to filename
    outPut.append(filename[0]) 
    
```

```python
trainDigitArrayNP = np.array(digitArrayList)
#This is training X

trainOutPutNP = np.array(outPut)
#This is traing Y
```

```python
print "training X shape: ", trainDigitArrayNP.shape
print "training Y shape: ", trainOutPutNP.shape
print "testing X shape: ", testDigitArrayNP.shape
print "testing Y shape: ", testOutPutNP.shape
```

```python
#Data has been prepared. Now it's time to clasify data
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
x = trainDigitArrayNP
y = trainOutPutNP
knn.fit(x, y)
```
